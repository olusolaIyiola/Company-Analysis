```python
import pandas as pd
import requests

# Replace with your Namsor API key
API_KEY = 'YOUR_NAMSOR_API_KEY'

# Load the names from the Excel file
df = pd.read_excel('Names_List.xlsx')

# Function to call Namsor API for a single name and save the response to a file
def call_namsor_api(firstname, lastname):
    url = f'https://api.namsor.com/onomastics/api/json/diaspora/batch?firstName={firstname}&lastName={lastname}&apiKey={API_KEY}'
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        with open(f'output/{firstname}_{lastname}.txt', 'w') as file:
            file.write(str(data))

# Loop through each row of the DataFrame and call Namsor API for each name
for index, row in df.iterrows():
    firstname = row['First Name']
    lastname = row['Last Name']
    call_namsor_api(firstname, lastname)
    print(f'Response for {firstname} {lastname} saved.')

# Function to extract 'ethnicity' and 'ethnicityAlt' fields from the response
def extract_ethnicity(response):
    if 'ethnicity' in response:
        return response['ethnicity']
    else:
        return None

def extract_alternate_ethnicity(response):
    if 'ethnicityAlt' in response:
        return response['ethnicityAlt']
    else:
        return None

# Read the saved responses and extract 'ethnicity' and 'ethnicityAlt' fields
for index, row in df.iterrows():
    firstname = row['First Name']
    lastname = row['Last Name']
    file_path = f'output/{firstname}_{lastname}.txt'
    
    with open(file_path, 'r') as file:
        data = eval(file.read())

    df.at[index, 'Ethnicity'] = extract_ethnicity(data)
    df.at[index, 'Alternate Ethnicity'] = extract_alternate_ethnicity(data)

# Save the updated DataFrame to a new Excel file
df.to_excel('Output_Names_List.xlsx', index=False)
print('Output file created successfully.')
```
